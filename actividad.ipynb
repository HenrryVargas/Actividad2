```python
import pandas as pd
```

```python
datos53EquinoccioPrimavera= pd.read_csv("53EquinoccioPrimavera\RME_402151514RMD00530000000_______P9.csv", header=None)
datos196SolsticioVerano = pd.read_csv("196SolsticioVerano\RME_414846256RMD01960000000_______P7.csv", header=None)
datos350EquinoccioOtono = pd.read_csv("350EquinoccioOtono\RME_428517516RMD03500000000_______P6.csv", header=None)
datos543SolsticioInvierno = pd.read_csv("543SolsticioInvierno\RME_445650976RMD05430000000_______P4.csv", header=None)

```

# 53 Equinoccio de primavera

```python

# Equinoccio de Primavera
df53EquiPrima = pd.DataFrame()
df53EquiPrima['LMST']= datos53EquinoccioPrimavera[1].str.replace("00053M","")
df53EquiPrima['LMST'] = pd.to_datetime(df53EquiPrima['LMST'],format="%H:%M:%S.%f")
df53EquiPrima['PAtmof']= datos53EquinoccioPrimavera[37]

# set the index to be columns "LMST" and drop=True to avoid creating a new index
df53EquiPrima.set_index('LMST', drop=True, inplace=True)
df53EquiPrima.DatetimeIndex = df53EquiPrima.index.time # remove the date part
dfEquiPrima = df53EquiPrima.resample('5T').mean()
dfEquiPrima
```

```python
# Grafica de equinoccio de primavera

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(dfEquiPrima.index, dfEquiPrima['PAtmof'], color='blue', label='Equinoccio de Primavera')

# set x begin with 00:00 and end with 23:05
ax.set_xlim([dfEquiPrima.index[0], dfEquiPrima.index[-1]])
ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=30))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.set_xlabel('Tiempo (Hora:Minutos)')
# x label are overlapping, so rotate the labels
fig.autofmt_xdate()

ax.set_ylabel('PresiÃ³n AtmosfÃ©rica (mbar)')
ax.set_title('Equinoccio de Primavera')
ax.grid(True)
plt.show()
```
